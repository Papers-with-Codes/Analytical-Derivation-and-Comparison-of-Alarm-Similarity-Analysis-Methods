## Implementation of F_2d
Note that this implemention can be improved in terms of response speed.

```python
import numpy as np
```

```python
# init config: can be changged and it affects accuracy
delta = 0.005
MAX1 = 5
MAX2 = 5
```

```python
def G(h1,h2,ro):
    return (1/(2*pi*sqrt(1-ro**2)))*exp(-(h1*h1+h2*h2-2*ro*h1*h2)/(2*(1-ro**2)))
```

```python
def F_2d(x_a,x_b,y_a,y_b,ro):
    
    h1=list()
    h2=list()

    # limit the bounds 
    if(x_a < -MAX1):
        x_a = -MAX1

    if(x_b < -MAX1):
        x_b = -MAX1

    if(y_a < -MAX1):
        y_a = -MAX1

    if(y_b < -MAX1):
        y_b = -MAX1


    if(x_a > MAX1):
        x_a = MAX1

    if(x_b > MAX1):
        x_b = MAX1

    if(y_a > MAX1):
        y_a = MAX1

    if(y_b > MAX1):
        y_b = MAX1

    h1=np.arange(x_a, x_b, delta).tolist()
    h2=np.arange(y_a, y_b, delta).tolist()

    h1, h2 = np.meshgrid(h1, h2)
    f = np.vectorize(G)
    return sum(sum(f(h1,h2,ro)))*delta*delta
```
