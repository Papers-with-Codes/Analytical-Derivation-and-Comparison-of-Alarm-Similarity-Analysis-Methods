## How to Generate a Covariance Matrix

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

This link shows the propertises of covariance matrix https://en.wikipedia.org/wiki/Covariance_matrix#Properties

```python
# covariance matrix generator, dim: size of square
def gen_cov(dim):
    n = dim
    a = 3
    np.random.seed
    A = np.matrix([np.random.randn(n) + np.random.randn(1)*a for i in range(n)])
    A = A*np.transpose(A)
    D_half = np.diag(np.diag(A)**(-0.5))
    C = D_half*A*D_half

    vals = list(np.array(C.ravel())[0])
    Ro = np.squeeze(np.asarray(C))
    return vals, Ro
```

```python
vals, Ro = gen_cov(dim = 15)
```

```python
# show histogram of random generated covariance matrix values
plt.hist(vals, range=(-1,1))
plt.show()
```

```python
# show covariance matrix as a hotmap
ax = sns.heatmap(Ro, vmin=-1, vmax=1 ,linewidth=0.01,  center=0, cmap='jet')
```

```python
# clustering the hot map
ax = sns.clustermap(Ro, row_cluster=True, col_cluster=True ,cmap="jet" ,linewidth=0.01)
```
